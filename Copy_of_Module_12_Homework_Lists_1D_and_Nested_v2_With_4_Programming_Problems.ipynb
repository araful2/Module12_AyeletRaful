### F1. Shopping List Manager (1D list + functions) (Suggested)
Write a program that maintains a shopping list (a 1D list of strings).

**Required functions:**
- `add_item(items, item)`
- `remove_item(items, item)` (remove by value if present)
- `print_items(items)` (numbered list)
- `save_items(items, filename)` (one item per line)
- `load_items(filename)` (returns a list)

**Program behavior:**
Provide a simple text menu in a loop:
1) add  2) remove  3) view  4) save  5) load  6) quit
The list should update as the user makes choices.


````python
# TODO: Implement F1 here
items = []
filename = "groceries.txt"
def add_items(items, item):
  items.append(item)

def remove_items(items, item):
  if item in items:
    items.remove(item)
  else:
    print("Item not found")

def print_items(items):
  if not items:
    print("Shopping list is empty.")
    return
  for i, item in enumerate(items, 1):
    print(f"{i}) {item}")

def save_items(items, filename):
  with open(filename, "w") as f:
    for line in items:
      f.write(line + "\n")

def load_items(filename):
  loaded_items = []
  try:
    with open(filename, "r") as a:
      for line in a:
        loaded_items.append(line.strip())
    print(f"Loaded items from {filename}.")
    return loaded_items
  except FileNotFoundError:
    print(f"File '{filename}' not found. Starting with an empty list.")
    return []

while True:
  user_input = input("Please choose one of the options: 1)add 2) remove 3) view 4) save 5) load 6) quit ").strip().lower()

  if user_input == "1" or user_input == "add":
    adding = input("Which item would you like to add to the list? ")
    add_items(items, adding)
    print(f"'{adding}' added.")

  elif user_input == "2" or user_input == "remove":
    removing = input("Which item would you like to remove from the list? ")
    remove_items(items, removing)

  elif user_input == "3" or user_input == "view":
    print("\nCurrent Shopping List:")
    print_items(items)
    print()

  elif user_input == "4" or user_input == "save":
    save_items(items, filename)
    print(f"Items saved to {filename}.")

  elif user_input == "5" or user_input == "load":
    items = load_items(filename)

  elif user_input == "6" or user_input == "quit":
    print("Quitting program. Goodbye!")
    break
  else:
    print("Invalid option. Please choose a number from 1 to 6 or the corresponding word.")
````

### F2. Grade Statistics (1D list of numbers + file input)
You will read a file of grades (one number per line) into a 1D list, then compute statistics.

**Requirements:**
- Read `grades.txt` into a list `grades`
- Compute and print: count, min, max, average
- Create a new list `curved` that adds +5 to each grade (cap at 100)
- Write `curved` to `curved_grades.txt` (one per line)

**If you don't have a `grades.txt` file, create one in code first** (write 8–12 sample grades).


````python
# TODO: Implement F2 here
marks = "80, 95, 100, 75, 86, 92, 93, 94, 88"
with open("grades.txt", "w") as f:
    f.write(marks)
def edit_marks():
  grades = []
  with open("grades.txt", "r") as h:
    line = h.read()
    new_list = line.split(",")
    for grade in new_list:
      grades.append(int(grade.strip()))

    minimum = min(grades)
    maximum = max(grades)
    count = len(grades)
    average = sum(grades)/count
    print(f"minimum: {minimum} maximum: {maximum} average:{average:.2f} count: {count}")
    return grades

current_grades = edit_marks()

def add_to_marks(grades):
  curved = []
  for i in grades:
    if i + 5 <= 100:
      curved.append(i)
  with open("curved_grades.txt", "w") as c:
    for num in curved:
      c.write(str(num) + "\n")



add_to_marks(current_grades)
````

### F4. Seating Chart (nested list) (2D list + loops + functions)
Create a seating chart using a nested list (list of rows).

**Requirements:**
- Build a 4×5 chart filled with the string `"EMPTY"`
- Write a function `assign_seat(chart, row, col, name)` that places a name
- Write a function `print_chart(chart)` that prints each row neatly
- Assign at least 6 names to different seats
- Count how many seats are still `"EMPTY"` and print that number

**Optional extension:** Write the seating chart to a file (one row per line).


````python
# TODO: Implement F4 here
rows = 4
cols = 5
chart = []
for r in range(rows):
    row = []
    for c in range(cols):
        row.append("EMPTY")
    chart.append(row)

def assign_seat(chart, row, col, name):
    chart[row][col] = name


def print_chart(chart):
  print("\nSeating Chart:")
  for row_data in chart:
    for value in row_data:
      print(f"{value:<10}", end="")
    print()
  print()


assign_seat(chart, 0, 1, "Esther")
assign_seat(chart, 1, 1, "Sara")
assign_seat(chart, 2, 1, "Rebecca")
assign_seat(chart, 3, 1, "Rachel")
assign_seat(chart, 2, 3, "Leah")
assign_seat(chart, 0, 2, "Mary")

print_chart(chart)

empty_seats = 0
for row_data in chart:
    for value in row_data:
      if value == "EMPTY":
        empty_seats += 1
print(f"Number of EMPTY seats: {empty_seats}")
````
